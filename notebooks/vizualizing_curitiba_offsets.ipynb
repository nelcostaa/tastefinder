```python
from pathlib import Path
import folium
from folium.plugins import MarkerCluster

import pandas as pd
```

```python
BASE_DIR = Path.cwd().parent  # parent é a raiz
data_path = BASE_DIR / "data" / "processed" / "curitiba_bairros_offsets_2025-05.csv"

df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
lat, lon = df.iloc[0][['latitude', 'longitude']]
lat, lon
```

```python
m = folium.Map(location=[lat, lon], zoom_start=13)

folium.CircleMarker(
 location=[lat, lon],
 radius=50,
 popup="Curitiba",
 color="blue",
 fill=True,
 fill_color="blue",
).add_to(m)

m
```

```python
# Crie o mapa centrado em Curitiba
m = folium.Map(location=[-25.4284, -49.2733], zoom_start=12)

# Adicione o cluster
marker_cluster = MarkerCluster().add_to(m)

# Adicione marcadores ao cluster
for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row.get('bairro', None)
    ).add_to(marker_cluster)

m  # Exibe no notebook
```
